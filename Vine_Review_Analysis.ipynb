```python
# Import pandas
import pandas as pd
import numpy as np
```

```python
# Import CSC
vine_table = "vine_table.csv"

# Create DF
vine_df = pd.read_csv(vine_table)
vine_df.head()
```

```python
# filter for total_votes >= 20
high_total_votes_df = vine_df[(vine_df["total_votes"] >= 20)]
high_total_votes_df.head()
```

```python
# retrieve rows where helpful_votes / total_votes >= 50%
helpful_votes_df = high_total_votes_df[(high_total_votes_df["helpful_votes"] / high_total_votes_df["total_votes"] >= .50)]
helpful_votes_df.head()
```

```python
# create a new df for vine reviewers (vine == "Y")
vine_reviews_df = helpful_votes_df[(helpful_votes_df["vine"] == "Y")]
vine_reviews_df.head()
```

```python
# create a new df for non-vine reviewers (vine == "N")
non_vine_reviews_df = helpful_votes_df[(helpful_votes_df["vine"] == "N")]
non_vine_reviews_df.head()
```

```python
# find total number of reviews, number of 5 start reviews, and % of 5 star reviews for vine reviewers
total_vine_reviews = vine_reviews_df.groupby(["vine"]).count()["review_id"]
five_star_vine_reviews = vine_reviews_df[(vine_reviews_df["star_rating"] == 5.0)].count()["review_id"]
percent_five_star_vine_reviews = (five_star_vine_reviews / total_vine_reviews)*100

vine_reviews_summary_df = pd.DataFrame({
   "Total Vine Reviews": total_vine_reviews,
   "Five Star Vine Reviews": five_star_vine_reviews,
   "Percent of Five Star Reviews": percent_five_star_vine_reviews})

vine_reviews_summary_df
```

```python
# find total number of reviews, number of 5 start reviews, and % of 5 star reviews for non-vine reviewers
total_nonvine_reviews = non_vine_reviews_df.groupby(["vine"]).count()["review_id"]
five_star_nonvine_reviews = non_vine_reviews_df[(non_vine_reviews_df["star_rating"] == 5.0)].count()["review_id"]
percent_five_star_nonvine_reviews = round((five_star_nonvine_reviews / total_nonvine_reviews)*100, 2)

nonvine_reviews_summary_df = pd.DataFrame({
   "Total Non-Vine Reviews": total_nonvine_reviews,
   "Five Star Non-Vine Reviews": five_star_nonvine_reviews,
   "Percent of Five Star Reviews": percent_five_star_nonvine_reviews})

nonvine_reviews_summary_df
```
